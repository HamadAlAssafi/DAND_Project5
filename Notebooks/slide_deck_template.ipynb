## Investigation Overview

**This presentation shows the most user type, who of these users will finish the trip faster, and why the finish it faster.**

## Dataset Overview

**GoFord bike System data about the bike trips in April for 2019, with some attributes such as users type and duration of trips and more. Data already cleaned in the wrangling phase.**

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df_clean = pd.read_csv('clean_file.csv')
df_clean.head()
```

> Note that the above cells have been set as "Skip"-type slides. That means

## What are the most user type

**The chart shows what are the most user type that we have in our dataset, and the most user type we have is the (Subscribers)**

```python
df_color = sb.color_palette()[0]
sb.countplot(data = df_clean, x = 'user_type', color = df_color);
plt.xlabel('User Type');
plt.title('Number of customer users and subsrciber users');
```

## What type of users is faster on finishing the trip

**This plot shows who are the user type that finishes the trip faster, and it is looks like the (subsrcibers) is faster than the (Customers)**

```python
sb.violinplot(data = df_clean, x = 'user_type', y = 'duration_sec',color = df_color, inner = None);
plt.title('The difference in duration per second between the (Subscribers) and (Customers)', size = 10);
plt.xlabel('User Type')
plt.ylabel('Duration Per Second')
```

## Why the subscribers is faster than the customers
**In this graphs we will see why the subscribers finishes the trip faster, and the most frequent latitude for the (Subscribers) is (65), and for the the (Customers) is (155),so that why the subscribers is faster than the customers**

```python
fig = plt.figure(figsize = [8,6]);
ax = sb.pointplot(data = df_clean, x = 'start_station_latitude', y = 'duration_min', hue = 'user_type',
           palette = 'Blues', linestyles = '');
plt.title('Biking durations with the start station latitude')
plt.xlabel('start station latitude')
plt.ylabel('Duration in minutes')
plt.yscale('log')
plt.yticks([1, 2, 5, 10, 20, 50, 100], [1, 2, 5, 10, 20, 50, 100]);
plt.xticks([0,10,20,35,50,65,80,95,110,125,140,155,170,185,200,215,230,245],[0,10,20,35,50,65,80,95,110,125,140,155,170,185,200,215,230,245]);
ax.set_yticklabels([],minor = True);
```

## Why the subscribers is faster than the customers
**In this graphs we will see why the subscribers finishes the trip faster,and the most frequent longitude for the (Subscribers) is between (215) and (230), and for the the (Customers) is between (170) and (185), so that why the subscribers is faster than the customers.**

```python
ax = sb.pointplot(data = df_clean, x = 'start_station_longitude', y = 'duration_min', hue = 'user_type',
           palette = 'Blues', linestyles = '');
plt.title('Biking durations with the start station longitude')
plt.xlabel('start station longitude')
plt.ylabel('Duration in minutes')
plt.yscale('log')
plt.yticks([1, 2, 5, 10, 20, 50, 100], [1, 2, 5, 10, 20, 50, 100]);
plt.xticks([0,10,20,35,50,65,80,95,110,125,140,155,170,185,200,215,230,245],[0,10,20,35,50,65,80,95,110,125,140,155,170,185,200,215,230,245]);
ax.set_yticklabels([],minor = True);
```
